```python
class Perceptron:
    def __init__(self, N):
        # ัะพะทะดะฐัั ะฝัะปะตะฒัะต ะฒะตัะฐ
        self.w = list()
        for i in range(N):
            self.w.append( 0)
            # ะผะตัะพะด ะฒััะธัะปะตะฝะธั ะทะฝะฐัะตะฝะธั ะฟะตััะตะฟััะพะฝะฐ
            
    def calc(self, x):
        res = 0
        for i in range(len(self.w)):
            res = res + self.w[i] * x[i]
        return res
    
    # ะฟะพัะพะณะพะฒะฐั ััะฝะบัะธั ะฐะบัะธะฒะฐัะธะธ ะฟะตััะตะฟััะพะฝะฐ 
    
    def sign(self, x):
        if self.calc(x)> 0:
            return 1
        else:
            return -1
        
    # ะพะฑััะตะฝะธะต ะฝะฐ ะพะดะฝะพะผ ะฟัะธะผะตัะต
    def learn(self, la, x, y):
        # ะพะฑััะฐะตะผ ัะพะปัะบะพ ะบะพะณะดะฐ ัะตะทัะปััะฐั ะฝะตะฒะตัะฝัะน
        if y * self.calc(x) <= 0:
            for i in range(len(self.w)):
                self.w[i] = self.w[i] + la * y * x[i]
    
    # ะพะฑััะตะฝะธะต ะฟะพ ะฒัะตะผ ะดะฐะฝะฝัะผ T - ะบะพััะตะถ ะฟัะธะผะตัะพะฒ
    def learning(self, la, T):
        # ัะธะบะป ะพะฑััะตะฝะธั 
        for n in range(100):
            # ะพะฑััะตะฝะธะต ะฟะพ ะฒัะตะผ ะฝะฐะฑะพัั ะฟัะธะผะตัะพะฒ
            for t in T:
                self.learn(la,t[0],t[1])
            
#ะ ัััะพะบะต 25 ะผั ะพัััะตััะฒะปัะตะผ ะบะพััะตะบัะธัะพะฒะบั ะฒะตัะพะฒ. ะะพัะผะพััะธะผ, ะบะฐะบ ััะธััั ะธ ัะฐะฑะพัะฐะตั ะฝะฐั ะฟะตััะตะฟััะพะฝ.

# ัะพะทะดะฐะตะผ ะบะปะฐัั ะดะฒัะผะตัะฝะพะณะพ ะฟะตััะตะฟััะพะฝะฐ

perceptron = Perceptron(2)
la = 0.1 # ะบะพัะฝัะฐัะฐ ะพะฑััะตะฝะธั

#ัะพะทะดะฐะตะผ ะฟัะธะผะตัั

T = list()
T.append([[2,1],1])
T.append([[3,2],1])
T.append([[4,1],1])
T.append([[1,2],-1])
T.append([[2,3],-1])
T.append([[5,7],-1])

perceptron.learning(la, T)# ะพะฑััะตะฝะธะต ะฟะตััะตะฟััะพะฝะฐ

print(perceptron.w) # ะฟะตัะฐัะฐะตะผ ะฒะตัะฐ
# ะฟัะพะฒะตัะธะผ ัะฐะฑะพัั ะฝะฐ ัะตััะพะฒัั ะฟัะธะผะตัะฐั

print(perceptron.sign([1.5,2]))

print(perceptron.sign([3,1.5]))

print(perceptron.sign([5,1]))

print(perceptron.sign([5,10]))
```

```python
import numpy as np
def sigmoid(x):
    # ััะฝะบัะธั ะฐะบัะธะฒะฐัะธะธ: f(x) = 1/(1 + e^(-x))
    return 1 / (1 + np.exp(-x))
class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)
    
weights = np.array([0,1]) # w1 = 0, w2 = 1
bias = 4                  # c = 4
n = Neuron(weights, bias)
x = np.array([2,3])       # x = 2, y = 3
print(n.feedforward(x))   #0.9990889488055994
```

ะะฐะบ ะพะฑััะฝะพ ัะฐะทะดะตะปะธะผ ะฝะฐะฑะพั ะดะฐะฝะฝัั ะฝะฐ ะดะฒะต ัะฐััะธ:

โ ัะตััะพะฒัะต ะดะฐะฝะฝัะต, ะฟะพ ะบะพัะพััะผ ะฑัะดะตั ะฟัะพะฒะตัััััั ัะพัะฝะพััั 
ะพะฑััะตะฝะฝะพะน ะผะพะดะตะปะธ.

ะั ัะฐะบะถะต ะธัะฟะพะปัะทัะตะผ seed(random_state=123) ั train_test_split, ััะพะฑั ะผั 
ะฒัะตะณะดะฐ ะฟะพะปััะฐะปะธ ะพะดะฝะพ ะธ ัะพ ะถะต ัะฐะทะดะตะปะตะฝะธะต ะธ ะผะพะณะปะธ ััะฐะฒะฝะธะฒะฐัั ะธ

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_digits, Y_digits, train_size = 0.80, test_size = 0.20, stratify = Y_digits, random_state = 123)

print('Train/Test Sizes : ', X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

ะะปั ะฝะฐัะฐะปะฐ ะฝะฐััะตะฝะธััะตะผ ะผะพะดะตะปั MLPClassifier ั ะฟะฐัะฐะผะตััะฐะผะธ ะฟะพ ัะผะพะปัะฐะฝะธั 
ะดะปั ััะตะฝะธัะพะฒะพัะฝัั ะดะฐะฝะฝัั.


```python
from sklearn.neural_network import MLPClassifier

mlp_classifier = MLPClassifier(random_state=123)
mlp_classifier.fit(X_train, Y_train)

MLPClassifier(activation = 'relu', alpha = 0.0001, batch_size = 'auto', beta_1=0.9,
             beta_2 = 0.999, early_stopping = False, epsilon = 1e-08,
             hidden_layer_sizes = (100,), learning_rate ='constant',
             learning_rate_init=0.001, max_iter=200, momentum = 0.9,
             n_iter_no_change=10, nesterovs_momentum=True, power_t=0.5,
             random_state=123, shuffle=True, solver='adam', tol=0.0001,
             validation_fraction=0.1, verbose=False, warm_start=False)
```

```python
Y_preds = mlp_classifier.predict(X_test)

print(Y_preds[:15])
print(Y_test[:15])

## ะะตัะพะด Score ะดะปั ะพัะตะฝะบะธ ัะพัะฝะพััะธ ะผะพะดะตะปะตะน ะบะปะฐััะธัะธะบะฐัะธะธ.
print('Test Accuracy : %.3f'%mlp_classifier.score(X_test,Y_test))

print('Training Accuracy : %.3f'%mlp_classifier.score(X_train,Y_train))
```

Cะพะทะดะฐะดะธะผ ะผะตัะพะด plot_confusion_matrix(), ะบะพัะพััะน ะฟัะธะฝะธะผะฐะตั ะธััะพะดะฝัะต ะธ

ะธัะฟะพะปัะทัั matplotlib.


```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(Y_test, Y_preds):
    conf_mat = confusion_matrix(Y_test, Y_preds)
    #print(conf_mat)
    fig = plt.figure(figsize=(6,6))
    plt.matshow(conf_mat, cmap=plt.cm.Blues, fignum=1)
    plt.yticks(range(10),range(10))
    plt.xticks(range(10),range(10))
    plt.colorbar();
    for i in range(10):
        for j in range(10):
            plt.text(i-0.2,j+0.1,str(conf_mat[j,i]),color = 'tab:red')
```

```python
plot_confusion_matrix(Y_test, mlp_classifier.predict(X_test))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn


def plot_confusion_matrix(Y_test, Y_preds):
    conf_mat = confusion_matrix(Y_test, Y_preds)
    print(conf_mat)
    fig = plt.figure(figsize=(5,5))
    plt.matshow(conf_mat, cmap=plt.cm.Blues, fignum=1)
    plt.yticks(range(3),['Setosa','Versicolor','Virginica'])
    plt.xticks(range(3),['Setosa','Versicolor','Virginica'])
    plt.colorbar();
    for i in range(3):
        for j in range(3):
            plt.text(i,j,str(conf_mat[j,i]))

# ะะฐะณััะทะบะฐ ะดะฐะฝะฝัั
data_url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
iris_data = pd.read_csv(data_url)

# ะะฐะทะดะตะปะตะฝะธะต ะดะฐะฝะฝัั ะฝะฐ ะฟัะธะทะฝะฐะบะธ ะธ ะผะตัะบะธ
X = iris_data.iloc[:, :4]  # ะัะธะทะฝะฐะบะธ
y = iris_data.iloc[:, 4]   # ะะตัะบะธ

# ะะฐะทะดะตะปะตะฝะธะต ะดะฐะฝะฝัั ะฝะฐ ะพะฑััะฐัััั ะธ ัะตััะพะฒัั ะฒัะฑะพัะบะธ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# ะกะพะทะดะฐะฝะธะต ะธ ะพะฑััะตะฝะธะต ะผะพะดะตะปะธ MLPClassifier ั ัะฒะตะปะธัะตะฝะฝัะผ ะบะพะปะธัะตััะฒะพะผ ะธัะตัะฐัะธะน
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=123)
mlp_classifier.fit(X_train, y_train)

# ะัะตะฝะบะฐ ะฟัะพะธะทะฒะพะดะธัะตะปัะฝะพััะธ ะผะพะดะตะปะธ
accuracy = mlp_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
#print(mlp_classifier.predict(X_test))
plot_confusion_matrix(y_test, mlp_classifier.predict(X_test))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neural_network import MLPRegressor

data_url = 'https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv'
salary_data = pd.read_csv(data_url)

# ะะฐะทะดะตะปะตะฝะธะต ะดะฐะฝะฝัั ะฝะฐ ะฟัะธะทะฝะฐะบะธ ะธ ะผะตัะบะธ
X = salary_data.iloc[:, :1]  # ะัะธะทะฝะฐะบะธ
y = salary_data.iloc[:, 1]   # ะะตัะบะธ

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)


mlp_regressor = MLPRegressor(hidden_layer_sizes=(100,100), max_iter=10000, random_state=12300)

mlp_regressor.fit(X_train, y_train)

y_preds = mlp_regressor.predict(X_test)

print(y_preds)
print(y_test)

print('Test R^2 Score : %.3f'%mlp_regressor.score(X_test, y_test))
print('Training R^2 Score : %.3f'%mlp_regressor.score(X_train,y_train))

print("Loss : ", mlp_regressor.loss_)

print("Number of Coefs : ", len(mlp_regressor.coefs_))
#[weights.shape for weights in mlp_regressor.coefs_]

print("Number of Iterations for Which Estimator Ran : ", mlp_regressor.n_iter_)

print("Name of Output Layer Activation Function : ", mlp_regressor.out_activation_)

```
