–≥–µ–Ω–∞–º–∏ x –∏ y. –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π Z. 
–ü—Ä–∏ —Ä–∞–≤–Ω–æ–º –∫–∞—á–µ—Å—Ç–≤–µ —Ö—Ä–æ–º–æ—Å–æ–º –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏–µ –æ—Ç–¥–∞–µ—Ç—Å—è —Ö—Ä–æ–º–æ—Å–æ–º–µ —Å 
–±–æ–ª—å—à–∏–º –Ω–æ–º–µ—Ä–æ–º. –ù–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ —Ö—Ä–æ–º–æ—Å–æ–º–∞ a —Å –≤—ã—Å—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º 
–ø–æ—Ä–æ–∂–¥–∞–µ—Ç —á–µ—Ç—ã—Ä–µ –Ω–æ–≤—ã—Ö —Ö—Ä–æ–º–æ—Å–æ–º—ã ùëè1, ùëê1, ùëè2, ùëê2, –æ–±–º–µ–Ω–∏–≤–∞—è—Å—å –≥–µ–Ω–∞–º–∏ —Å

–ü–æ—Å–ª–µ–¥–Ω—è—è —Ö—Ä–æ–º–æc–æ–º–∞ (—Å –Ω–∏–∑—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º) –≤—ã–±—ã–≤–∞–µ—Ç –∏–∑ 
–ø–æ–ø—É–ª—è—Ü–∏–∏. –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –≤ 
–ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏.

–ü–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ù–∞–ø–∏—à–µ–º –∏—Ö.


–†–µ—à–µ–Ω–∏–µ:
–ù–∞—á–Ω–µ–º —Å —Ñ—É–Ω–∫—Ü–∏–∏ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã qZ(x,y):

```python
def qZ (x, y):
    return (x-3 * y + 1) / (3 * x ** 2 +3 * y ** 2 +1)
# –î–∞–ª–µ–µ, –æ—Ü–µ–Ω–∏–º —Å—É–º–º–∞—Ä–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ö—Ä–æ–º–æ—Å–æ–º:

def qSumZ (Z):
    return sum(Z)


def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range (4)]
    Y = [0 for i in range(4)]

    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]

    X[0] = oldX[sortedId[0]]

    X[1] = oldX[sortedId[1]]

    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]

    Y[2] = oldY[sortedId[0]]

    Y[3] = oldY[sortedId[1]]

    return X, Y

#–û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞—à–∏—Ö –ø–æ—Ç–æ–º–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã:

def sorting(Z):
    sortedId = sorted(range(len(Z)),key = lambda k: Z[k])

    return sortedId

#–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —à–∞–≥–∞ —ç–≤–æ–ª—é—Ü–∏–∏:
#—à–∞–≥ —ç–≤–æ–ª—é—Ü–∏–∏

def evoStep(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = Z[:]

    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)

    return X,Y,Z

#–ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —ç–≤–æ–ª—é—Ü–∏–æ–Ω–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∑–∞–¥–∞—á–µ–π - 4 —à–∞–≥–∞

#—à–∞–≥–∏ —ç–≤–æ–ª—é—Ü–∏–∏ (–∫–æ–Ω–µ—á–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è), –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 4 —à–∞–≥–∞

def evoSteps(X, Y, stepsNum = 4):
    results = []

    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]

        X, Y, Z = evoStep(X,Y, arrZ)

        X, Y = exchangeScheme(X,Y,sorting(Z))

        results.append([X,Y,qSumZ(arrZ),arrZ])
    return X,Y,results


```

–Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –≥–µ–Ω–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ (–æ—Ü–µ–Ω–∫–∏ 
–∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º, —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø–æ—Ç–æ–º–∫–æ–≤ –∏ —ç–≤–æ–ª—é—Ü–∏–æ–Ω–Ω—ã—Ö —à–∞–≥–æ–≤), —Ä–µ—à–∏–º 
–∑–∞–¥–∞—á—É –≤ —á–∏—Å–ª–∞—Ö. –ü—É—Å—Ç—å –¥–∞–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –º–∞—Å—Å–∏–≤—ã —Ö—Ä–æ–º–æ—Å–æ–º X –∏ Y:

–ó–∞–ø–∏—à–µ–º –∏—Ö –≤ —Ç—Ä–µ–±—É–µ–º–æ–º –≤–∏–¥–µ –∏ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –Ω–∞–ø–∏—Å–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π 
evoSteps.

```python
X = [-2,-1,0,1]
Y = [-2,-1,0,1]

results = evoSteps(X,Y)

```

```python
def qZ(x, y):
    return (x + 3*y) / (x ** 2 + y ** 2 + 1)
# –î–∞–ª–µ–µ, –æ—Ü–µ–Ω–∏–º —Å—É–º–º–∞—Ä–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ö—Ä–æ–º–æ—Å–æ–º:

def qSumZ(Z):
    return sum(Z)


def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]

    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]

    X[0] = oldX[sortedId[0]]

    X[1] = oldX[sortedId[1]]

    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]

    Y[2] = oldY[sortedId[0]]

    Y[3] = oldY[sortedId[1]]

    return X, Y

#–û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞—à–∏—Ö –ø–æ—Ç–æ–º–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã:

def sorting(Z):
    sortedId = sorted(range(len(Z)),key = lambda k: Z[k])

    return sortedId

#–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —à–∞–≥–∞ —ç–≤–æ–ª—é—Ü–∏–∏:
#—à–∞–≥ —ç–≤–æ–ª—é—Ü–∏–∏

def evoStep(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = Z[:]

    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)

    return X,Y,Z

#–ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —ç–≤–æ–ª—é—Ü–∏–æ–Ω–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∑–∞–¥–∞—á–µ–π - 4 —à–∞–≥–∞

#—à–∞–≥–∏ —ç–≤–æ–ª—é—Ü–∏–∏ (–∫–æ–Ω–µ—á–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è), –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 4 —à–∞–≥–∞

def evoSteps(X, Y, stepsNum = 4):
    results = []

    for i in range(10):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]

        X, Y, Z = evoStep(X,Y, arrZ)

        X, Y = exchangeScheme(X,Y,sorting(Z))

        results.append([X,Y,qSumZ(arrZ),arrZ])
    return X,Y,results

X = [-5,-3,-2,-1]
Y = [-1,-2,0,1]

results = evoSteps(X,Y)



for i in range(len(results[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')

qualityArrZ = []
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]

print(f'max Z: {max(qualityArrZ)}')
```

–¢–µ–ø–µ—Ä—å, –Ω–∞–ø–∏—à–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ, —Å–æ–≥–ª–∞—Å–Ω–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –≤—ã—à–µ, —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è 
—Ä–∞–±–æ—Ç—ã –º–µ—Ç–æ–¥–∞ –∏–º–∏—Ç–∞—Ü–∏–∏ –æ—Ç–∂–∏–≥–∞. –í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É —Ä–µ–±—Ä–∞:

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞

def edgeLength(i,j,distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] ==i and item[0] ==j) else -1) for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])
    
#–í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É –º–∞—Ä—à—Ä—É—Ç–∞:
def routeLength(V, distances):
    edges = []
    
    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i],V[i + 1], distances))
        
    return sum(edges)

#–ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–¥–Ω–æ–∫—Ä–∞—Ç–Ω–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏:

# –æ–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏

def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0
    
    arrV[Z[0] - decrement], arrV[Z[1]-decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]
    
    return arrV

#–ê —Ç–µ–ø–µ—Ä—å —Ñ—É–Ω–∫—Ü–∏—é, –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ —Å–∞–º–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏:
# –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V
```

–¢–µ–ø–µ—Ä—å, –æ–ø–∏—à–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –≤—ã–±–æ—Ä–∞ –ø–æ–¥—Ö–æ–¥—è—â–µ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞:

```python
# –≤—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances)  # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength]  # –º–∞—Å—Å–∏–æ–≤ —Å—É–º–º –¥–ª–∏–Ω
    
    myBestS = sumLength

    myBestV = V

    # —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i])  # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances)  # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength  # —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞

        if myBestS > newS:
            myBestV = newV
            myBestS = newS

        # –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –±–ª–∏–Ω–∞–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T)  # –ø–æ–¥—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

            # –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ p, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è

            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T)  # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã

    return myBestV, arrSum
```

```python
# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É 

def drawRouteGraph(distances, bestRoute):

    newDistances = []

# –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1): 
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)
                
    graph = nx.Graph() #—Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph. add_weighted_edges_from(newDistances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä 
    
# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)
```

–ò, –Ω–∞–∫–æ–Ω–µ—Ü, —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –Ω–∞–∏–ª—É—á—à–∏–π –º–∞—Ä—à—Ä—É—Ç –∏ –µ–≥–æ –¥–ª–∏–Ω—É

```python
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}') 
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances, bestRoute) # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```

### –ó–∞–¥–∞–Ω–∏–µ
–ù–∞–π—Ç–∏ –¥–ª–∏–Ω—É –≥–∞–º–∏–ª—å—Ç–æ–Ω–æ–≤–∞ —Ü–∏–∫–ª–∞ S4 –≤ –ø–æ–ª–Ω–æ–º –≥—Ä–∞—Ñ–µ K6 –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö 
—Ü–∏–∫–ª–æ–≤ —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞ –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞–º –Ω–∏–∂–µ


```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:

import networkx as nx
from math import e

distances = [(1,2,22),(1,3,43),(1,4,39),(1,5,28),(1,6,20),
             (2,3,26),(2,4,33),(2,5,36),(2,6,17),(3,4,22),(3,5,40),(3,6,24),
             (4,5,22),(4,6,19),(5,6,20)]        # –¥–ª–∏–Ω—ã —Ä–µ–±–µ—Ä

V = [1,3,4,5,6,2,1]     # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3,4),(4,6),(5,2),(6,2)]       # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P = [51,23,29,31]       # C–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞ –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á–µ—Ç–∞

T = 100     # –Ω–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞

# –ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã:

# —Ñ—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)
# —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—ã

def reductTemp(prevT):
    nextT = 0.5 * prevT

    return nextT

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –≤–µ—Ä—à–∏–Ω–∞–º, –∫–∞–∫ –≤ –Ω–∞—á–∞–ª—å–Ω–æ–º —É—Å–ª–æ–≤–∏–∏:

graph = nx.Graph()      #—Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
graph.add_weighted_edges_from(distances)        #–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏

nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)


# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞

def edgeLength(i, j, distances, roundTrip=True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in
                    distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])


# –í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É –º–∞—Ä—à—Ä—É—Ç–∞:
def routeLength(V, distances):
    edges = []

    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))

    return sum(edges)


# –ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–¥–Ω–æ–∫—Ä–∞—Ç–Ω–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏:

# –æ–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏

def routeOneReplacement(arrV, Z, replacementByName=True):
    decrement = 1 if replacementByName else 0

    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]

    return arrV


# –ê —Ç–µ–ø–µ—Ä—å —Ñ—É–Ω–∫—Ü–∏—é, –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ —Å–∞–º–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏:
# –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V


# –≤—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances)  # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength]  # –º–∞—Å—Å–∏–æ–≤ —Å—É–º–º –¥–ª–∏–Ω
    
    myBestS = sumLength

    myBestV = V

    # —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i])  # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances)  # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength  # —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞

        if myBestS > newS:
            myBestV = newV
            myBestS = newS

        # –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –±–ª–∏–Ω–∞–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T)  # –ø–æ–¥—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

            # –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ p, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è

            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T)  # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã

    return myBestV, arrSum


# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É

def drawRouteGraph(distances, bestRoute):
    newDistances = []

    # –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and \
                    distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph()  # —Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances)  # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

    # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color='#fb7258', node_size=2000, with_labels=True)
```

```python
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances, bestRoute) # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```

–ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:

import networkx as nx
from math import e

distances = [(1,2,22),(1,3,43),(1,4,39),(1,5,28),(1,6,20),
             (2,3,26),(2,4,33),(2,5,36),(2,6,17),(3,4,22),(3,5,40),(3,6,24),
             (4,5,22),(4,6,19),(5,6,20)]        # –¥–ª–∏–Ω—ã —Ä–µ–±–µ—Ä

V = [1,3,4,5,6,2,1]     # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3,4),(4,6),(5,2),(6,2),(5,3),(4,5),(2,3),(5,6),(3,4),(4,6),(5,2),(6,2),(5,3),(4,5),(2,3),(5,6)]       # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P = [51,23,29,31,29,31,29,31,29,31,29,31,29,31,29,31]       # C–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞ –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á–µ—Ç–∞

T = 100     # –Ω–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞

# –ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã:

# —Ñ—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)
# —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—ã

def reductTemp(prevT):
    nextT = 0.8 * prevT

    return nextT

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –≤–µ—Ä—à–∏–Ω–∞–º, –∫–∞–∫ –≤ –Ω–∞—á–∞–ª—å–Ω–æ–º —É—Å–ª–æ–≤–∏–∏:

graph = nx.Graph()      #—Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
graph.add_weighted_edges_from(distances)        #–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏

nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)


# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞

def edgeLength(i, j, distances, roundTrip=True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in
                    distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])


# –í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É –º–∞—Ä—à—Ä—É—Ç–∞:
def routeLength(V, distances):
    edges = []

    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))

    return sum(edges)


# –ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–¥–Ω–æ–∫—Ä–∞—Ç–Ω–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏:

# –æ–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏

def routeOneReplacement(arrV, Z, replacementByName=True):
    decrement = 1 if replacementByName else 0

    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]

    return arrV


# –ê —Ç–µ–ø–µ—Ä—å —Ñ—É–Ω–∫—Ü–∏—é, –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ —Å–∞–º–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏:
# –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V


# –≤—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances)  # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength]  # –º–∞—Å—Å–∏–æ–≤ —Å—É–º–º –¥–ª–∏–Ω
    
    myBestS = sumLength

    myBestV = V

    # —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i])  # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances)  # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength  # —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞

        if myBestS > newS:
            myBestV = newV
            myBestS = newS

        # –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –±–ª–∏–Ω–∞–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T)  # –ø–æ–¥—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

            # –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ p, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è

            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T)  # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã

    return myBestV, arrSum


# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É

def drawRouteGraph(distances, bestRoute):
    newDistances = []

    # –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and \
                    distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph()  # —Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances)  # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

    # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color='#fb7258', node_size=2000, with_labels=True)
```

```python
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances, bestRoute) # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```
